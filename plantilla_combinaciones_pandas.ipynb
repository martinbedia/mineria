## Merge

```python
left = pd.DataFrame(
       {
           "key": ["K0", "K1", "K2", "K3"],
           "A": ["A0", "A1", "A2", "A3"],
           "B": ["B0", "B1", "B2", "B3"],
       }
   )

right = pd.DataFrame(
       {
           "key": ["K0", "K1", "K2", "K3"],
           "C": ["C0", "C1", "C2", "C3"],
           "D": ["D0", "D1", "D2", "D3"],
       }
   )
```

```python
left
```

```python
right
```

```python
pd.merge(left,right, on="key")    # on especificamos la columna
```

```python
left = pd.DataFrame(
       {
           "key": ["K0", "K1", "K2", "K3"],
           "A": ["A0", "A1", "A2", "A3"],
           "B": ["B0", "B1", "B2", "B3"],
       }
   )

right = pd.DataFrame(
       {
           "key": ["K1", "K0", "K2", "K3"],
           "C": ["C1", "C0", "C2", "C3"],
           "D": ["D1", "D0", "D2", "D3"],
       }
   )
```

```python
pd.merge(left,right, on= "key")   # al unirlo por key cambian los valores que estaban desordenados, comparar con el anterior
# busca las tablas que se corresponden con los mismos valores que se quieren juntar
```

```python
left = pd.DataFrame(
    {
        "key1": ["K0", "K0", "K1", "K2"],
        "key2": ["K0", "K1", "K0", "K1"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)

right = pd.DataFrame(
    {
        "key1": ["K0", "K1", "K1", "K2"],
        "key2": ["K0", "K0", "K0", "K0"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)
```

```python
left
```

```python
right
```

```python
pd.merge(left,right, on= ["key1","key2"])   # how= 'inner'
# mira las claves para las que hay valores en ambas tablas
```

Ademas de inner podemos utilizar los siguientes modificadores:
- left
- right (no se usa casi)
- outer
- inner
- cross

El how se suele poner antes del on.

```python
pd.merge(left,right, how= 'outer',on= ["key1","key2"])
```

```python
# el cross es el unico que no lleva claves (se usa poco)
# junta las filas sin criterio ninguno
```

```python
df1 = pd.DataFrame({"col1": [0,1], "col_left": ["a","b"]})
df2 = pd.DataFrame({"col1": [1,2,2], "col_right": [3,4,5]})

pd.merge(df1,df2, on="col1", how= 'outer', indicator= True)
```

## Join

```python
left = pd.DataFrame(
       {"A": ["A0", "A1", "A2"], "B": ["B0", "B1", "B2"]}, index=["K0", "K1", "K2"]
   )

right = pd.DataFrame(
       {"C": ["C0", "C2", "C3"], "D": ["D0", "D2", "D3"]}, index=["K0", "K2", "K3"]
   )
```

```python
left
```

```python
right
```

A diferencia del merge que es una  funcion de pandas y entonces la invocamos en la forma pd.merge(), el join es un metodo de los objetos DataFrame y entonces se invoca df.join. El df tenemos que pensar que es nuestra tabla izquierda

```python
left.join(right)  # añade los valores en aquellos lugares donde lo tiene
# seria un equivalente a un left join entre left y right
```

```python
right.join(left)
```

```python
left.join(right, how= "outer")    # casi equivalente a poner right.join(left), mezcla todas con todas
```

```python
left.join(right, how= "inner")
```

```python
# intentar no hacer esto
df= pd.merge(left,right, left_index = True, right_index= True, how = "inner")   # mismo de arriba pero peor sintaxis
df
```
